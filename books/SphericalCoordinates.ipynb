# Spherical coordinates

This book will briefly plot an example for the functioning of the `some_tools.spherical_conversion.py` module.
Once installed the `some_tools` library, just import the module and the class `SphericalStream`


```python
%matplotlib inline

from obspy import read
from some_tools.spherical_conversion import SphericalStream

# Example with standard ObsPy stream
st = read()
st.detrend('linear')
st.filter('bandpass', freqmin=0.2, freqmax=20.0)
st.detrend('linear')
st.detrend('constant')
_ = st.plot(equal_scale=False)
```

Now we convert and plot the out-stream

```python
# Convert
sphe_obj = SphericalStream(st, convert=True)
sphe_st = sphe_obj.get_converted_stream()

# Plot
_ = sphe_st[:3].plot(equal_scale=False)
_ = sphe_st[3:].plot(equal_scale=False)

st_all = st + sphe_st[:3]
_ = st_all.plot(equal_scale=False)
_ = sphe_st[3:].plot(equal_scale=False)
```
